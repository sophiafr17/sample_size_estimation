# Goal
improve conversion rate from home to the second page in the sales funnel

# Assume:
1. product manager has told us that we only care about an improvement of at least +1%. Baseline is 10%

```python
import pandas as pd
import statsmodels.stats.api as sms
import numpy as np
import matplotlib.pyplot as plt
```

## Estimate Sample Size

```python

p1_p2 = sms.proportion_effectsize(0.1,0.11)

sample_size = sms.NormalIndPower().solve_power(p1_p2, 
                                               power = 0.8,
                                               alpha = 0.05)

print("The required sample size per group is:", '\n',
      round(sample_size)
      )
```

## Effect Size vs. Sample Size

```python
p2_list = np.arange(.105,.155,0.005)
print(p2_list)
```

```python
sample_size = []

for i in p2_list:
    p1_p2 = sms.proportion_effectsize(0.1,i)
    sample_size.append( round(sms.NormalIndPower().solve_power(p1_p2,
                                                   power = 0.8,
                                                   alpha = 0.05
                                                   )))
    
print(sample_size)
```

```python
plt.plot(p2_list,sample_size)
plt.xlabel('possible_p2 value')
plt.ylabel('sample size')
```
